Used Crazy Eights ruleset:

- 2 players, indexed 0 and 1, player 0 goes first
- Each player start with 7 cards
- Players discard cards which suit or rank matches the top card of the discard pile, or eights
- If a player plays an eight, the next player must follow suit or play another eight
- If a player can't play, they must draw up to 3 cards from the drawing deck until they are able to play. If they're still unable to play after drawing 3 cards, they must pass
- If the drawing deck is exhausted, all the discarded cards except the top card is reshuffled to form a new drawing deck
